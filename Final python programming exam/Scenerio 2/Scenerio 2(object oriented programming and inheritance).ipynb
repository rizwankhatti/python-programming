```python

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print(f"{self.year} {self.make} {self.model}'s engine started.")

    def stop_engine(self):
        print(f"{self.year} {self.make} {self.model}'s engine stopped.")

    def display_info(self):
        print(f"Vehicle: {self.year} {self.make} {self.model}")


class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def display_info(self):
        super().display_info()
        print(f"Type: Car | Number of Doors: {self.doors}")

    def accelerate(self):
        print(f"{self.make} {self.model} is accelerating smoothly.")


class Truck(Vehicle):
    def __init__(self, make, model, year, capacity):
        super().__init__(make, model, year)
        self.capacity = capacity

    def display_info(self):
        super().display_info()
        print(f"Type: Truck | Cargo Capacity: {self.capacity} tons")

    def load_cargo(self, weight):
        if weight <= self.capacity:
            print(f"Loading {weight} tons of cargo onto {self.make} {self.model}.")
        else:
            print(f"Cannot load {weight} tons. Exceeds capacity!")

   
class Motorcycle(Vehicle):
    def __init__(self, make, model, year, drive_type):
        super().__init__(make, model, year)
        self.drive_type = drive_type

    def display_info(self):
        super().display_info()
        print(f"Type: Motorcycle | Drive Type: {self.drive_type}")

    def do_wheelie(self):
        print(f"{self.make} {self.model} is doing a wheelie!")


if __name__ == "__main__":
    
    car1 = Car("Toyota", "Corolla", 2022, 4)
    truck1 = Truck("Volvo", "FH16", 2021, 20)
    bike1 = Motorcycle("Yamaha", "R1", 2023, "Chain Drive")

  
    car1.start_engine()
    truck1.start_engine()
    bike1.start_engine()

    print("\n--- Vehicle Information ---")
    car1.display_info()
    truck1.display_info()
    bike1.display_info()

    print("\n--- Specific Actions ---")
    car1.accelerate()
    truck1.load_cargo(15)
    truck1.load_cargo(25)  
    bike1.do_wheelie()

    car1.stop_engine()
    truck1.stop_engine()
    bike1.stop_engine()
```
