1. task: give me the sum of the values passed as unknown args to function **args**
2. task: give me the sum of the values passed as unknown args to function as **kwargs**

```python
# 1. task: give me the sum of the values passed as unknown args to function **args**

def add_nums(name,*nums):
    result = 0
    for num in nums:
        result += num
    print(f"{name} : {result}")

add_nums("Rizwan", 2,3,4,5,6,7)
    
```

```python
# 1. task: give me the sum of the values passed as unknown args to function **args**

def add_nums(name,*nums):
    print(f"{name} : {sum(nums)}")

add_nums("Rizwan", 2,3,4,5,6,7)
```

```python
# 2. task: give me the sum of the values passed as unknown args to function as **kwargs**

def add_nums(name,**nums):
    result = 0
    for val in nums.values():
        result += val
    print(f"{name} : {result}")

add_nums("Rizwan",num1=2,num3=3,num4=4,num5=5,num6=6,num7=7)
```

```python
# 2. task: give me the sum of the values passed as unknown args to function as **kwargs**

def add_nums(name,**nums):
    print(f"{name} : {sum(nums.values())}")

add_nums("Rizwan",num1=2,num3=3,num4=4,num5=5,num6=6,num7=7)
```

  - first bracket
  - second  by using variable

```python
# first bracket

(lambda x,y:x+y) (8,9)
```

```python
# second by using variable

l = (lambda x,y:x+y)
```

```python
def f(n):
    if n==1:
        return 1
    s = n+f(n-1)
    return s
f(4)
```

```python
def f(n):
    if n==1:
        return 1
    s = n*f(n-1)
    return s
f(4)
```

```python
f = lambda n: 1 if n == 1 else n*f(n-1)
f(4)
```
