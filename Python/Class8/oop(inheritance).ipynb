## Inheritance

```python
class parents():

    def __init__(self, eye_color, hair_color):
        self.eye_color = eye_color
        self.hair_color = hair_color

    def can_paint(self):
        print(f"have ability to draw/paint arts")
```

```python
par_obj = parents("brown", "brown")
#------------------------------
print(par_obj.eye_color)
print(par_obj.hair_color)
par_obj.can_paint()
```

```python
class child(parents):
    pass
```

```python
child_obj = child()
```

```python
child_obj = child("brown", "black")
#---------------------------
print(child_obj.eye_color)
print(child_obj.hair_color)
child_obj.can_paint()
```

```python
print(par_obj.hair_color)
print(child_obj.hair_color)
```

```python
# additional characteristics in child class

class child(parents):

    def __init__(self, eye_color, hair_color, skin_color):
        super().__init__(eye_color, hair_color) # for single class inhertance
        # Father.__init__() # for multi class inhertance
        # Mother.__init__() # for multi class inhertance
        self.skin_color = skin_color

    def can_drive(self):
        print(f"child can drive as well.")

# jub ap additonal characteristics ko add krty ho tw apny class ko batana pharta hai k baki characteristics hain wo kahan jaen gai tw usky liye ik special keyword ata hai that is super()
# ye super keyword aysy hai jesy self keyword class aur object ki internal managment k liye use krty hain wasy hi super ka keyword python manage krta hai for inheritance.
```

```python
child_obj = child("brown", "black","brown")
#---------------------------
print(child_obj.eye_color)
print(child_obj.hair_color)
print(child_obj.skin_color)
child_obj.can_paint()
```

TASK:

- create a class with name "Father" which have two attributes and one method.
- create a class with name "Mother" which have one attribute.
- create a class with namke "Child" that inherit characteris of "Father" and "Mother" but should ahve atleast two more attribute and methods other than "father class"

```python
class father():
    
    def __init__(self, eye_color, hair_color):
        self.eye_color = eye_color
        self.hair_color = hair_color

    def can_paint(self):
        print(f"have ability to draw/paint arts")

class mother():
    
    def __init__(self, driving):
        self.driving = driving

    def driving(self):
        print(f"mother has skill of {self.driving} driving")
        
class child(father, mother):
    
    def __init__(self, eye_color, hair_color, driving, programming): # constructor overriding
        father.__init__(self, eye_color, hair_color)
        mother.__init__(self, driving)
        self.programming = programming

    def can_sing(self):
        print(f"child can sing aswell")
    
```

```python
father_obj = father('green','white')
```

```python
print(father_obj.eye_color)
print(father_obj.hair_color)
father_obj.can_paint()
```

```python
mother_obj = mother('bmw')
```

```python
print(mother_obj.driving)
```

```python
child_obj = child('Brown','Black','Car','python programming')
```

```python
print(child_obj.eye_color)
print(child_obj.hair_color)
print(child_obj.driving)
print(child_obj.programming)
child_obj.can_paint()
child_obj.can_sing()
child_obj.driving()

```

```python
print(father_obj.eye_color)
print(child_obj.eye_color)
```

## method overriding

```python
class animals():
    def eating(self):
        print("animals can eat")

class birds():
    def eating(self): # method overriding
        print("birds can eat")
```

```python
ani_obj = animals()
ani_obj.eating()
```

```python
birds_obj = birds()
birds_obj.eating()
```

## overloading

```python
from typing import overload

class adder():
    
    @overload
    def add(self, x:int, y:int) -> int:
        ...

    @overload
    def add(self, x:float, y:float) -> float:
        ...
    @overload
    def add(self, x:str, y:str) -> str:
        ...

    # Actual Functionality
    def add(self, x, y):
        if isinstance(x,int) and isinstance(y,int):
            return x+y
        elif isinstance(x,float) and isinstance(y,float):
            return x+y
        elif isinstance(x,str) and isinstance(y,str):
            return x+y
        else:
            print("TYPE ERROR")
    
```

```python
adder_obj = adder()
```

```python
adder_obj.add(2,8)
```

```python
adder_obj.add(2.5,8.0)
```

```python
adder_obj.add("Rizwan", "Rehman")
```
